```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
#we need to use plt.show in .py files
plt.show()
```

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_axes([0,0,1,1])
ax.set_title("Title")
ax.set_xlabel('x')
yticks = [0, 50, 100, 150,200]
ax.set_yticks(yticks)
ax.set_ylabel('y')
ax.plot(x,y)
```
